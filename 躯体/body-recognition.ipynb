```python
import os
from mmpose.apis import MMPoseInferencer
import torch
torch.backends.cudnn.benchmark = True
```

```python
# 2. 指定输入视频路径 (替换此处)
video_input_path = r"D:\视频标注\交大-顾凯\顾凯-2.mp4" # <--- 修改这里

# 3. 指定输出目录
vis_output_dir = r"D:\视频标注\交大-顾凯\顾凯-躯体\顾凯-2-vis"      # 保存可视化视频的目录
results_output_dir = r"D:\视频标注\交大-顾凯\顾凯-躯体\顾凯-2-results"  # 保存结果数据 (例如 JSON) 的目录

# 4. 指定设备 ('cuda:0' 代表第一个 GPU, 'cpu' 代表 CPU)
device = 'cuda:0' 

# --- 用户配置结束 ---

# 打印配置信息以供确认
print(f"使用设备: {device}")
print(f"输入视频: {video_input_path}")
# print(f"使用检测器别名: {det_model_alias if 'det_model_alias' in locals() else '未指定或使用路径'}")
# print(f"使用 2D 姿态别名: {pose2d_model_alias if 'pose2d_model_alias' in locals() else '未指定或使用路径'}")
# print(f"使用 3D 姿态别名: {pose3d_model_alias if 'pose3d_model_alias' in locals() else '未指定或使用路径'}")
print(f"可视化输出目录: {vis_output_dir}")
print(f"结果数据输出目录: {results_output_dir}")

# 基本检查输入视频是否存在
if not os.path.exists(video_input_path):
    print(f"\n!!!  在 {video_input_path} 未找到输入视频 !!!")
else:
    print(f"\n输入视频已找到。")
```

```python
#  创建输出目录

os.makedirs(vis_output_dir, exist_ok=True)
os.makedirs(results_output_dir, exist_ok=True)

print(f"已确保输出目录存在: {vis_output_dir}")
print(f"已确保输出目录存在: {results_output_dir}")
```

```python
#初始化推理器：使用别名
inferencer = MMPoseInferencer(
    pose2d="rtmo",
    device = device
 ) # 使用之前定义的设备 ('cuda:0' 或 'cpu')
```

```python
# 单元格 5: 对视频运行推理
# 这会调用 inferencer。它返回一个生成器。
# 实际的处理、可视化保存和结果转储将在你迭代生成器时（下一个单元格）发生。

print(f"开始对视频进行推理处理: {video_input_path}")
print(f"可视化结果将保存到: {vis_output_dir}")
print(f"结果数据将保存到: {results_output_dir}")

# 对于视频输入，inferencer 返回一个生成器对象
results_generator = inferencer(
    inputs=video_input_path,
    batch_size=16,
    vis_out_dir=vis_output_dir,
    out_dir=results_output_dir,
    # 你可以探索的可选参数 (默认值通常即可):
    # show=False,             # 在无头环境中不要尝试显示窗口
    # draw_bbox=True,         # 在可视化中绘制边界框
    # draw_heatmap=False,     # 绘制 2D 热图 (如果适用)
    # skeletal_style='mmpose' # 可视化风格
)

print("Inferencer 已调用。处理将在迭代结果生成器时开始。")
```

```python
# 单元格 6: 处理结果 (触发执行和保存)
# 你必须迭代生成器，以使 inferencer 处理视频帧、保存可视化结果并转储结果数据。

print("正在迭代结果生成器以处理视频帧...")
print("(根据视频长度和硬件情况，这可能需要相当长的时间)")

# 迭代生成器以触实际发处理
# 我们不一定需要在这里将所有结果存储在列表中，
# 因为它们正被保存到由 vis_out_dir 和 out_dir 指定的文件中。
frame_count = 0
for i, result in enumerate(results_generator):
    # 你可以在这里选择性地检查单帧的 'result' 字典内容
    # print(f"已处理第 {i} 帧")
    frame_count = i + 1
    # 如果需要，可以添加进度指示
    if frame_count % 30 == 0:
         print(f"  已处理 {frame_count} 帧...")


print(f"\n视频处理完毕。总共处理了 {frame_count} 帧。")
```

```python
import json

# --------------- 配置区 ---------------
INPUT_PATH  = r"D:\视频标注\交大-杨一凡-待筛选\杨一凡-躯体\杨一凡-5-results\predictions\杨一凡-5.json"  # 原始文件
OUTPUT_PATH = r"D:\视频标注\交大-杨一凡-待筛选\杨一凡-躯体\杨一凡-5-results\predictions\杨一凡-5-tonly.json"  # 过滤后保存路径
SCORE_THRESH = 0.6                  # 有效关键点阈值
# ---------------------------------------

def filter_best_instance_per_frame(data, thresh):
    filtered = []
    for frame in data:
        instances = frame.get('instances', [])
        if not instances:
            # 如果没有检测到任何人，就原样保留一个空 instances 列表
            filtered.append({'frame_id': frame['frame_id'], 'instances': []})
            continue

        # 统计每个实例中 keypoint_scores > thresh 的数量
        counts = [sum(1 for s in inst['keypoint_scores'] if s > thresh)
                  for inst in instances]
        # 找到数量最多的实例下标
        best_idx = max(range(len(counts)), key=lambda i: counts[i])

        # 保留该实例
        best_inst = instances[best_idx]
        # 按原格式组织
        filtered.append({
            'frame_id': frame['frame_id'],
            'instances': [best_inst]
        })

    return filtered

def main():
    # 读取原始 JSON
    with open(INPUT_PATH, 'r', encoding='utf-8') as f:
        data = json.load(f)

    # 过滤
    filtered_data = filter_best_instance_per_frame(data, SCORE_THRESH)

    # 写回 JSON
    with open(OUTPUT_PATH, 'w', encoding='utf-8') as f:
        json.dump(filtered_data, f, ensure_ascii=False, indent=2)

    print(f'过滤完成，结果已保存到 {OUTPUT_PATH}')

if __name__ == '__main__':
    main()
```
